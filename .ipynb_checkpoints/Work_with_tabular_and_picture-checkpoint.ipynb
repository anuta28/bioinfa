```python
%%R
data_long <- pivot_longer(data, cols = c(T_count, A_count, C_count, G_count), names_to = "Letter", values_to = "Count")
class(data_long)

data_long <- data_long %>%
  group_by(Letter) %>%
  mutate(Sequence_Index = row_number()) %>%
  ungroup()

ggplot(data_long, aes(x = Sequence_Index, y = Count, color = Letter, group = Letter)) +
  geom_line() +
  geom_point() +
  labs(
    x = "Sequence Index",
    y = "Count",
    title = "Counts of T, A, C, G in Sequences",
    color = "Letter"
  ) +
  theme_minimal()
```
