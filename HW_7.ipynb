Видим, что на этом филогенетическом дереве прослеживается эволюция (накопление мутаций) в S-белке коронавируса. Штаммы, полученные в Ухане, наиболее близки, немного дальше отстоит штамм из Индии. Видно, что в целом последовательость белка консервативна. Однако, накопленные со временем эволюции мутации позволяют построить филогенетическое дерево, которое отражает современные взгляды на таксономию крупных групп животных. Эволюционно моложе млекопитающие - верблюд, свинья, корова, собака. Сестринскими группами стали приматы и грызуны - человек и серая крыса.
