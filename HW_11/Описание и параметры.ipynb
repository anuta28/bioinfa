## Задание 11 (Оптимизация сборки генома de novo)

В этом задании сравнивались 3 различных сборщика: Platanus, SPAdes, Megahit.

Platanus не поддерживает передачу множественной передачи k-mer, поэтому можно лишь подобрать один оптимальный или собрать в общий файл разные результаты - для отдельных k-merов результаты плохие, общая длина сборки маленькая, параметры качества неудовлетворительные, сборка такие файлов в один хоть и увеличивает общую длину, но поднимает лишь количественные, но не качественные параметры. Такой сборщик подходит вероятно, для довольно коротких геномов.

SPAdes - более автоматизированная и удобная вещь, можно передать список к-меров, а также ограничивающие параметры по длине и качеству контигов, в итоге результат лучше, чем у platanus

Megahit - наиболее подхдящий сборщик для прочтений Illumina по результатам этой работы. Есть довольно большое количество параметров, отвечающих за аккуратность сборки (например, уровень пуринов)- видим, что данный сборщик лучше всего справился по показателям как качества, так и количества на довольно небольшом наборе параметров, что можно наблюдать в quast отчете.

megahit -1 /projects/mipt_dbmp_biotechnology/genome_de_novo/7_S4_L001_R1_001.fastq -2 /projects/mipt_dbmp_biotechnology/genome_de_novo/7_S4_L001_R2_001.fastq -o megahit_output --k-list 21,33,55,77 --min-count 2 --prune-level 2 -t 8 --memory 0.8
