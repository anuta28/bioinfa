## Домашнее задание 4

В этом задании предлагается проанализировать последовательности гомологичных белков из 4х разных биологических видов. Для этого необходимо провести множественное выравнивание при помощи одного из онлайн тулов с сайта EBI и построить филогенетическое дерево для этих белков.

В качестве рассматриваемого белка выбран белок цитохром с, так как он является распространенным среди самых разных организмов. Это небольшой гем-содержащий белок, который играет ключевую роль в процессе клеточного дыхания. Он участвует в переносе электронов в дыхательной цепи переноса электронов (ЭТЦ) в митохондриях. Цитохром c является частью комплекса III (цитохром bc1-комплекс) и комплекса IV (цитохром c-оксидаза) дыхательной цепи. Также цитохром с связывает энергетический метаболизм и регуляцию клеточной смерти, что делает его объектом активных исследований в биохимии, молекулярной биологии и медицине.
